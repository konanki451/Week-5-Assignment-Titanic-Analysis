```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
data = sns.load_dataset('titanic')

# Convert survived to categorical with labels
data['survived'] = data['survived'].map({0: 'No', 1: 'Yes'})


# Bar plot of Survival count

plt.figure()
counts = data['survived'].value_counts()
plt.bar(counts.index, counts.values, color=['red', 'green'])
plt.title('Frequency Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()


```

```python
# Histogram of Age

plt.figure()
plt.hist(data['age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

```

```python
# Boxplot of Age by Sex

plt.figure()
sns.boxplot(x='sex', y='age', data=data)
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
