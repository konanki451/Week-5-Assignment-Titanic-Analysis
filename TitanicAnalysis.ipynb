```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
data = pd.read_csv("Titanic.csv")

# Convert Survived to clear labels
data['Survived'] = data['Survived'].map({0: 'No', 1: 'Yes'})

# 1. Bar plot  Survival counts

fig, ax = plt.subplots(figsize=(7, 5))

counts = data['Survived'].value_counts()
ax.bar(counts.index, counts.values, color=['#d9534f', '#5cb85c'], edgecolor='black')

ax.set_title('Frequency Distribution of Survival')
ax.set_xlabel('Survived')
ax.set_ylabel('Number of Passengers')

fig.tight_layout()
fig.savefig('Bar_chart.png', dpi=200)
plt.show()

# 2. Histogram  Age distribution

fig, ax = plt.subplots(figsize=(7, 5))

ax.hist(data['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')

ax.set_title('Age Distribution of Passengers')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Passengers')

fig.tight_layout()
fig.savefig('Histogram.png', dpi=200)
plt.show()

# 3. Boxplot  Age by Sex

fig, ax = plt.subplots(figsize=(7, 5))

sns.boxplot(x='Sex', y='Age', data=data, ax=ax, palette='pastel')

ax.set_title('Age Distribution by Sex')
ax.set_xlabel('Sex')
ax.set_ylabel('Age')

fig.tight_layout()
fig.savefig('Boxplot.png', dpi=200)
plt.show()
```
