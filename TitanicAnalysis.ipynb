```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
data = pd.read_csv("Titanic.csv")

# Convert Survived to clear labels
data['Survived'] = data['Survived'].map({0: 'No', 1: 'Yes'})

# 1. Bar plot  Survival counts
plt.figure()

counts = data['Survived'].value_counts()
plt.bar(counts.index, counts.values)

plt.title('Frequency Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')

plt.tight_layout()
plt.savefig('Bar_chart.png', dpi=200)
plt.show()


# 2. Histogram  Age distribution
plt.figure()

plt.hist(data['Age'].dropna(), bins=20)

plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')

plt.tight_layout()
plt.savefig('Histogram.png', dpi=200)
plt.show()


# 3. Boxplot  Age by Sex
plt.figure()

# Prepare data grouped by sex
male_age = data[data['Sex'] == 'male']['Age'].dropna()
female_age = data[data['Sex'] == 'female']['Age'].dropna()

plt.boxplot([male_age, female_age], labels=['Male', 'Female'])

plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')

plt.tight_layout()
plt.savefig('Boxplot.png', dpi=200)
plt.show()
```
